```python
# Read data into numpy array
import numpy as np
data_file = 'spectrum/108_DATA.E'
with open(data_file, "rb") as f:
    data = np.fromfile(f, dtype=np.uint16)

data = data.reshape(2,-1)
```

```python
data.shape
```

```python
import matplotlib
matplotlib.use('qt5agg')
import matplotlib.pyplot as plt

%matplotlib inline

# Data for plotting
e = np.arange(1024)
y1 = data[0]
y2 = data[1]

fig, ((ax11, ax12)) = plt.subplots(2, 1, figsize=(15,10), dpi=80)
ax11.plot(e, y1, 'k.-')
ax11.set_title('First detector')
ax11.set_xlabel('Binding energy (eV)')
ax11.set_ylabel('Intensity (a.u.)')
ax11.grid()

ax12.plot(e, y2, 'k.-')
ax12.set_title('Second detector')
ax12.set_xlabel('Binding energy (eV)')
ax12.set_ylabel('Intensity (a.u.)')
ax12.grid()

fig.tight_layout()
plt.show()
```
